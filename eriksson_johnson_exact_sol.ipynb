```python
import matplotlib.pyplot as plt
import numpy as np
import math
from matplotlib import cm
from matplotlib.ticker import LinearLocator
```

```python
def get_r(eps):
    r1 = (1 + math.sqrt((1 + (4*(eps**2)*(math.pi**2)))))/(2*eps)
    r2 = (1 - math.sqrt((1 + (4*(eps**2)*(math.pi**2)))))/(2*eps)

    return r1, r2
```

```python
x = np.linspace(0, 1, 10000)
y = np.linspace(0, 1, 10000)
X, Y = np.meshgrid(x, y)
eps_list = [0.1, 0.01, 0.001]
Z = {}
for i, eps in enumerate(eps_list):
    r1, r2 = get_r(eps)
    Z[i] = np.sin(np.pi*Y)*(np.exp(r1*(X-1)) - np.exp(r2*(X-1)))/(np.exp(-r1) - np.exp(-r2))

# plt.contourf(X, Y, Z, 100, cmap='jet')
```
